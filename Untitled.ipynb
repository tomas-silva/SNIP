# C - Population distiction and Toomre diagram 

```python
final_gal_v = []
final_gal_uw = []

for i in range(len(gal_vel)):
    final_gal_v.append(gal_vel[i][1])
    final_gal_uw.append(np.sqrt(gal_vel[i][0]**2+gal_vel[i][2]**2))
```

```python
sigma_u_thin, sigma_v_thin, sigma_w_thin, V_asym_thin, X_thin = 35, 20, 16, -15, 0.85
sigma_u_thick, sigma_v_thick, sigma_w_thick, V_asym_thick, X_thick = 67, 38, 35, -46, 0.09
sigma_u_halo, sigma_v_halo, sigma_w_halo, V_asym_halo, X_halo = 160, 90, 90, -100, 0.0015
```

```python
def f_pops(U, V, W, pop):
    if pop == 1:
        sigma_u, sigma_v, sigma_w, V_asym = sigma_u_thin, sigma_v_thin, sigma_w_thin, V_asym_thin
    if pop == 2:
        sigma_u, sigma_v, sigma_w, V_asym = sigma_u_thick, sigma_v_thick, sigma_w_thick, V_asym_thick
    if pop == 3:
        sigma_u, sigma_v, sigma_w, V_asym = sigma_u_halo, sigma_v_halo, sigma_w_halo, V_asym_halo
        
    k = 1/(((2*np.pi)**(3/2))*sigma_u*sigma_v*sigma_w)
    
    f = k*np.exp(-(U**2)/(2*sigma_u**2) -((V-V_asym)**2)/(2*sigma_v**2) -(W**2)/(2*sigma_w**2) )
        
    return f
```

```python
final_pops = []
t_frac_thick = []
green_circles_x = []
green_circles_y = []
pop_list = []
pop_list_s2 = []
td_d = []

count_thin = 0
count_thick = 0
count_thick_s2 = 0
count_thin_s2 = 0
count_undecided_s2 = 0
count_halo = 0

P_halo_max = 0

for star_index in range(len(gal_vel)):
    pop = 0
    fi_pi_comp = 0
    
    total_prob = 0
    for i in range(3):
        i=i+1
        
        f_new = f_pops(gal_vel[star_index][0], gal_vel[star_index][1], gal_vel[star_index][2], i)
        
        if i == 1:
            X = X_thin
            
        if i == 2:
            X = X_thick
            
        if i == 3:
            X = X_halo
    
        fi_pi = X*f_new
        
        total_prob += fi_pi
        
               
        if fi_pi > fi_pi_comp:
            fi_pi_comp = fi_pi
            pop = i
    
    P_thin = X_thin*f_pops(gal_vel[star_index][0], gal_vel[star_index][1], gal_vel[star_index][2], 1)/total_prob
    P_thick = X_thick*f_pops(gal_vel[star_index][0], gal_vel[star_index][1], gal_vel[star_index][2], 2)/total_prob
    P_halo = X_halo*f_pops(gal_vel[star_index][0], gal_vel[star_index][1], gal_vel[star_index][2], 3)/total_prob
    
    if P_halo > P_halo_max:
        P_halo_max = P_halo
        
    td_d.append(P_thick/P_thin)
    
    if P_thick/P_thin > 2:
        green_circles_x.append(gal_vel[star_index][1])
        green_circles_y.append(np.sqrt(gal_vel[star_index][0]**2+gal_vel[star_index][2]**2))
        
        pop_list_s2.append('thick')
        
        count_thick_s2 += 1
        
        
    if P_thin/P_thick > 2:
        #green_circles_x.append(gal_vel[star_index][1])
        #green_circles_y.append(np.sqrt(gal_vel[star_index][0]**2+gal_vel[star_index][2]**2))
        count_thin_s2 += 1
        
        pop_list_s2.append('thin')
        
    if P_thick/P_thin < 2 and P_thin/P_thick < 2:
        pop_list_s2.append('undecided')
        count_undecided_s2 += 1
        
    t_frac_thick.append(P_thick)
    final_pops.append(pop)
    
    if pop == 1:
        count_thin += 1
        pop_list.append('thin')
    if pop == 2:
        count_thick += 1
        pop_list.append('thick')
    if pop == 3:
        pop_list.append('halo')
        count_halo += 1
```

```python
print('2 thin and thick disk pops')
print('Thin disk stars:', count_thin)
print('Thick disk stars:', count_thick)
print('')
print('Uncertain disk pops allowed')
print('Thin disk stars:', count_thin_s2)
print('Thin disk stars:', count_thick_s2)
print('Thin disk stars:', count_undecided_s2)
print('')
print('Halo stars:', count_halo)
print('Max probability for an halo stars:', P_halo_max)
```

```python
import matplotlib.pyplot as plt
cm = plt.cm.get_cmap('RdYlBu')

x = final_gal_v
y = final_gal_uw
z = td_d
sc = plt.scatter(x, y, c=z, s=10, cmap=cm)

plt.colorbar(sc)
plt.clim(0, 10)

for i in range(len(green_circles_x)):
    plt.scatter(green_circles_x[i], green_circles_y[i], s=80, facecolors='none', edgecolors='tab:green')

# Show the boundary between the regions:
r0 = -100
theta = np.arange(0, 4*np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta),'--', c='grey')

r0 = -200
theta = np.arange(0, 4*np.pi / 2, 0.01)

plt.plot(r0 * np.cos(theta), r0 * np.sin(theta),'--', c='grey')

axes = plt.gca()

axes.set_xlim([-200,50])

axes.set_ylim([-10,200])
plt.show()
```
